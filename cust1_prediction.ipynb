```python
import pandas as pd
import joblib
from preprocessing_utils import preprocess_churn_data
```

```python
xgb_model, scaler, encoders = joblib.load("xgb_churn_model.pkl")

new_data = pd.read_csv("customer_churn_dataset-testing-master.csv")
```

```python
new_data.shape
```

```python
new_data.head()
```

```python
X_new, _, _, _, customer_ids = preprocess_churn_data(
    new_data,
    is_train=False,
    scaler=scaler,
    encoders=encoders
)
```

```python
churn_pred = xgb_model.predict(X_new)

churn_prob = xgb_model.predict_proba(X_new)[:, 1]

results = pd.DataFrame({
    'CustomerID': customer_ids,
    'Predicted_Churn': churn_pred,
    'Churn_Probability': churn_prob
})

results.head()
```

```python
results.to_csv("cust1_churn_predictions.csv", index=False)
print("Predictions saved to churn_predictions.csv")
```
